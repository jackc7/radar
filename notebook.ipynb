```python
# Read and process data
with open("output.txt", 'r') as f:
    data = []
    for x, line in enumerate(f.readlines()):
        wl = line.strip()
        if len(wl) != 3:
            continue
        try:
            data.append(int(wl))
        except ValueError:
            pass
```

```python

data = data[8750:9500]
# data = data[3715:4715]
# for i, x in enumerate(data):
#     print(f"{i*100}\t{x}")
plt.plot(data)
plt.ylim(300, 800)
plt.ylabel('Voltage')
plt.show()
```

```python
SAMPLE_RATE = 2000

# Number of samples in normalized_tone
N = len(data)

yf = fft(data)
xf = fftfreq(N, 1 / SAMPLE_RATE)

plt.plot(xf, np.abs(yf))
plt.xlim(0,1000)
plt.ylim(0, 60000)
plt.show()

plt.plot(xf/27.34, np.abs(yf))
plt.xlim(0,40)
plt.ylim(0, 10000)
plt.show()
```

```python
# http://chris35wills.github.io/simple_fft_filter/
yfn = yf
xfn = xf
yfn[:13], yfn[2730:] = 0, 0

inv = ifft(yfn)
plt.plot(inv)
plt.xlim(0,40)
# plt.ylim(0, 60000)
plt.show(block=False)
```
